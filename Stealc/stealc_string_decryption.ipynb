```python
import base64
import pefile
```

```python
file_path = rb'/tmp/stealc_sample.bin'
data = open(file_path,'rb').read()
pe = pefile.PE(data=data)

for section in pe.sections:
    if section.Name.decode().strip('\x00\x00') == '.rdata':
        rdata_section = section
        break

if rdata_section:
    rdata_data = rdata_section.get_data()

rdata_data = rdata_data.split(b'\n')

for item in rdata_data:
    if b'==' in item:
        base64_data = item.split(b'\x00\x00\x00\x00')[2:-1]

my_list = [i.decode() for i in base64_data]
```

## Base64

```python
def base_decode(string):
    decoded = base64.b64decode(string)
    return decoded
```

## RC4 Decryption

```python

key = "52129722198130874989795557381261264814249348323986"


def rc4(data, key):
    if type(data) == str:
        data = data.encode('utf-8')
    if type(key) == str:
        key = key.encode('utf-8')

    j = 0
    s = list(range(256))
    for i in range(256):
        j = (j + s[i] + key[i % len(key)]) % 256
        s[i], s[j] = s[j], s[i]

    i = 0
    j = 0
    out = []
    for x in data:
        i = (i + 1) % 256
        j = (j + s[i]) % 256
        s[i], s[j] = s[j], s[i]
        out.append(x ^ s[(s[i] + s[j]) % 256])
    return bytes(out)
```
